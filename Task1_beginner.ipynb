## Control Structures

```python
# Check if a Number is Prime (Using for and if)

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

## Matplotlib Basics

```python
import matplotlib.pyplot as plt

# Simple line chart
plt.figure(figsize=(10, 5))
plt.plot(df['hp'],df.index , marker='o', color='b', label='hp')

# Labels and title
plt.xlabel("Horsepower ")
plt.ylabel("Car Index")
plt.title("Horsepower of Cars")
plt.grid(True)
```

## Data Aggregation

```python
# Group by 'cyl' and compute mean for 'mpg' and 'hp'
grouped_df = df.groupby("cyl")[["mpg", "hp"]].mean()

# Plotting grouped data
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot for MPG
grouped_df["mpg"].plot(kind="bar", ax=ax[0], color="skyblue")
ax[0].set_title("Average MPG by Cylinder")
ax[0].set_ylabel("Miles per Gallon (mpg)")

# Bar plot for HP
grouped_df["hp"].plot(kind="bar", ax=ax[1], color="salmon")
ax[1].set_title("Average Horsepower by Cylinder")
ax[1].set_ylabel("Horsepower (hp)")

# Show plots
plt.tight_layout()
plt.show()
```

## Data Cleaning

```python
sard = pd.read_csv("student_admission_record_dirty.csv")
sard.head()
```

```python
sard.info()
```

```python
sard.isna().sum()
```

```python
# Drop rows where multiple columns have missing values
sard_cleaned = sard.dropna(thresh=5)

# Fill missing categorical values
sard_cleaned["Name"].fillna("Unknown", inplace=True)
sard_cleaned["City"].fillna("Unknown", inplace=True)
sard_cleaned["Gender"].fillna(sard_cleaned["Gender"].mode()[0], inplace=True)
sard_cleaned["Admission Status"].fillna(sard_cleaned["Admission Status"].mode()[0], inplace=True)

# Fill missing numerical values with mean
sard_cleaned["Age"].fillna(sard_cleaned["Age"].mean(), inplace=True)
sard_cleaned["Admission Test Score"].fillna(sard_cleaned["Admission Test Score"].mean(), inplace=True)
sard_cleaned["High School Percentage"].fillna(sard_cleaned["High School Percentage"].mean(), inplace=True)

# Verify if missing values are handled
sard_cleaned.isnull().sum()
```

## Exploratory Data Analysis (EDA)

### Titanic Dataset

```python
titanic = pd.read_csv('tested.csv')
titanic.head()
```

```python
titanic.info()
```

The dataset contains 418 passengers, with the following key details:

- Target Variable: Survived (0 = No, 1 = Yes)

- Passenger Info: Age, Sex, Pclass (ticket class)

- Family Details: SibSp (siblings/spouse), Parch (parents/children)

- Ticket & Fare Details: Ticket, Fare, Cabin, Embarked

### Handling Missing data

```python
# Fill missing Age values with median age
titanic["Age"].fillna(titanic["Age"].median(), inplace=True)

# Fill missing Fare values with median fare
titanic["Fare"].fillna(titanic["Fare"].median(), inplace=True)

# Drop Cabin column
titanic.drop(columns=['Cabin'], inplace=True)    # Majorly missing values

# Verify missing values are handled
titanic.isnull().sum()
```

```python
# Set up figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Survival Count (Bar Plot)
sns.countplot(data=titanic, x="Survived", ax=axes[0, 0], palette="viridis")
axes[0, 0].set_title("Survival Count")
axes[0, 0].set_xticklabels(["Not Survived", "Survived"])

# 2. Survival by Gender (Bar Plot)
sns.countplot(data=titanic, x="Sex", hue="Survived", ax=axes[0, 1], palette="coolwarm")
axes[0, 1].set_title("Survival by Gender")

# 3. Age Distribution (Histogram)
sns.histplot(titanic["Age"], bins=20, kde=True, ax=axes[1, 0], color="blue")
axes[1, 0].set_title("Age Distribution")

# 4. Fare Distribution (Histogram)
sns.histplot(titanic["Fare"], bins=20, kde=True, ax=axes[1, 1], color="green")
axes[1, 1].set_title("Fare Distribution")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

The multiplot provides several insights into the Titanic dataset:

1) Survival Count (Top Left):

- The majority of passengers did not survive.

- The number of survivors is significantly lower than non-survivors.

2) Survival by Gender (Top Right):

- A higher proportion of females survived compared to males.

- Most male passengers did not survive.

3) Age Distribution (Bottom Left):

- The distribution is right-skewed, with most passengers between 20 and 40 years old.

- There are fewer passengers in the older age range.

4) Fare Distribution (Bottom Right):

- The fare distribution is highly right-skewed, meaning most passengers paid low fares.

- A few passengers paid significantly higher fares, indicating first-class travelers.
